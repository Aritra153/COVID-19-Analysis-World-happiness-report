```python
#Calculation of the maximum infection rates of all the countries(and creating a new column for it)
countries=list(agg_dataset.index)
max_infectrate=[]
for i in countries:
    max_infectrate.append(agg_dataset.loc[i].diff().max())
agg_dataset["Max_infection_rate"] = max_infectrate    
```

```python
#corona deaths dataset insertion
death_data=pd.read_csv("covid19_deaths_dataset.csv")
death_data
```

```python
#aggregating the dataset
agg_deathdata=death_data.drop(["Lat","Long","Province/State"],axis = 1).groupby("Country/Region").sum()
```

```python
countries1=list(agg_deathdata.index)
max_deathrate=[]
for i in countries1:
    max_deathrate.append(agg_deathdata.loc[i].diff().max())
agg_deathdata["Max_Death_rate"]=max_deathrate
```

```python
#Another new column containing the maximum death rate is added to the table
agg_deathdata
```

```python
#A new dataset containing only the Max_Death_rate column
new_deathdataagg=pd.DataFrame(agg_deathdata["Max_Death_rate"])
new_deathdataagg
```

```python
#Joining the two datasets containing the Max_infection_rate and Max_Death_rate columns only
dataset1=new_datasetagg.join(new_deathdataagg,how="inner")
```

```python
dataset1
```

```python
#plotting Max_Infection_rate vs Max_Death_rate
X=dataset1["Max_infection_rate"]
Y=dataset1["Max_Death_rate"]
sns.scatterplot(X,Y)
```

```python
#Using linear regression to get the best st.line plot
sns.regplot(X,Y)
```

```python
#importing the world happiness report file
happiness_report=pd.read_csv("worldwide_happiness_report.csv")
happiness_report
```

```python
#Selecting some of the useless columns out
useless_cols=["Overall rank","Score","Generosity","Perceptions of corruption"]
happiness_report.drop(useless_cols,axis=1)

```

```python
#aggregating the data
agg_happinessreport=happiness_report.drop(useless_cols,axis=1).groupby("Country or region").sum()
agg_happinessreport
```

```python
#Joining the max infection rate and max death rate with happiness report of the countries
combined_data=dataset1.join(agg_happinessreport,how="inner")
combined_data
```

```python
#using Linear regression to plot Max_infection_rate vs GDP per capita 
X1=combined_data["GDP per capita"]
Y1=combined_data["Max_infection_rate"]
sns.scatterplot(X1,Y1)
```

```python
#As the plot is not a good fit,I will plot X1 vs log(Y1)
X1=combined_data["GDP per capita"]
Y1=combined_data["Max_infection_rate"]
sns.scatterplot(X1,np.log(Y1))
```

```python
#Linear regression model
X1=combined_data["GDP per capita"]
Y1=combined_data["Max_infection_rate"]
sns.regplot(X1,np.log(Y1))
```

```python
#Similarly plotting Max_Death_rate vs GDP per capita
X2=combined_data["GDP per capita"]
Y2=combined_data["Max_Death_rate"]
sns.regplot(X2,np.log(Y2))
```

```python
#using Linear regression to plot Max_infection_rate vs Healthy life expectancy
X3=combined_data["Healthy life expectancy"]
Y3=combined_data["Max_infection_rate"]
sns.scatterplot(X3,np.log(Y3))
```

```python
#drawing the regression line
sns.regplot(X3,np.log(Y3))
```

```python
#using Linear regression to plot Max_infection_rate vs Healthy life expectancy
X4=combined_data["Healthy life expectancy"]
Y4=combined_data["Max_Death_rate"]
sns.scatterplot(X4,np.log(Y4)) #log scaling Y4 
```
